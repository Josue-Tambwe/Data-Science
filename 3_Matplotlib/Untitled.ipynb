La **Data Visualization** est une pratique transformant des données brutes en graphiques, diagrammes, etc...Dans le but de faciliter la lecture ainsi que la compréhension des données. Les bibliothèque **Matplotlib** et **Seaborn** permettent de visualiser les données en Python.

# Matplotlib : 
Génère des graphiques et complète Numpy et Pandas dans l'analyse des données.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

### On veut tracer le graphe de l'évolution de la concentration (*ordonnée*) d'un produit pharmaceutique dans le sang en faction du temps (*abscisse*)
Nous utiliserons deux tableaux uni-dimensionnels de même longueur (pour faire la correspondance une ordonnée pour une abscisse)

### Le nuage des points

```python
temps = np.arange(1, 11)
concentration = np.array([5.5, 7.2, 11, 13.6, 19, 21.8, 30.2, 33, 35.8, 36])
# Représentation sous forme de points avec 'Scatter'
plt.scatter(temps, concentration, marker = 'o')
# la méthode 'show()' affiche le graphique (mais est optionnelle pour le notebook Jupyter)
plt.show()
```

**Remarques** : le paramètre ***marker*** permet de spécifier l'apparence des points : 
- 'o' pour des points ronds comme des petits cercles
- ',' ou  's' pour des carrés (pixels)
- 'v' pour des triangles tournés vers le bas
- '+' pour avoir des '+'
- 'x' pour avoir des 'x'
-  etc...

### Les courbes

```python
plt.plot(temps, concentration)
plt.show()
```

On peut mettre en évidence les points d'intersections entre ordonnés et abscisses avec le paramètre **marker**

```python
plt.plot(temps, concentration, marker = 's')
plt.show()
```

On peut attribuer une couleur à la courbe représentative de la fonction avec le paramètre **color**

```python
plt.plot(temps, concentration, marker = 's', color = 'green')
plt.show()
```

On peut modifier l'apparence des traits de la courbe avec le paramètre **linestyle** : 
- continue ( ____ ) : **'solid'**
- petits traits discontinus (- - -) : **'dashed'**
- pointillés(....) : **'dotted'**

```python
plt.plot(temps, concentration, color = 'red', linestyle = 'dashed')
plt.show()
```

```python
plt.plot(temps, concentration, color = 'purple', linestyle = 'dotted')
plt.show()
```

On peut spécifier l'épaisseur des traits avec **linewidth**

```python
plt.plot(temps, concentration, color = 'yellow', linestyle = 'solid', linewidth = 3)
plt.show()
```

***Il existe encore d'autres paramètres, cependant ceux présentés suffisent pour tracer des courbes***

### Ajout des titres, grille et légende : 

 - tritre : plt.title('titre_du_graphique')
 - grille : plt.grid(True)
 - axe des abscisses : plt.xlabel('nom_axe_abscisse')
 - axe des ordonnées : plt.ylabel('nom_axe_ordonnee')
 - légence d'une courbe : avec le paramètre **label*, donner un nom à cette courbe puis avant l'affichage, utiliser **plt.legend()**

```python
x = np.linspace(0, 20, 50)
plt.plot(x, x*2, linestyle = 'solid', color = 'blue', label = 'f(x) = 2x')
plt.plot(x, x*15, linestyle = 'dashed', color = 'red', label = 'f(x) = 15x')
plt.plot(x, x**2, linestyle = 'dotted', color = 'green', label = 'f(x) = x²')
plt.grid(True)
plt.title("Mon Graphique")
plt.xlabel('Temps')
plt.ylabel('Concentration du produit dans le sang')
plt.legend()
plt.show()
```

## Représenter graphiquement les courbes des fonctions mathématiques (ou autres)

```python
x = np.linspace(-10, 10, 100)
# La fonction cosinus
plt.plot(x, np.cos(x), linestyle = 'solid', label = 'cosinus(x)')
plt.plot(x, np.sin(x), linestyle = 'dashed', color = 'orange', label = 'sinus(x)')
plt.xlabel('abscisse(x)')
plt.title('Représentation graphique de la fonction Sinus et Cosinus')
plt.legend()
plt.show()
```

# Les Graphiques 3D : 
Avec mpl_toolkits, qui est un sous module de Matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Création de la figure
fig = plt.figure(figsize = (8,8))
# Ajout des axes à la figure 
axe = fig.add_subplot(1, 1, 1, projection = '3d')
# Génération des données aléatoires (20 nombres pour chaque vecteur)
X = np.random.randint(-10, 10, 20)
Y = np.random.randint(-10, 10, 20)
Z = np.random.randint(-10, 10, 20)
# Dessin du nuage des points
axe.scatter(X, Y, Z, color = "red", marker = "o")
# Titres des axes
axe.set_xlabel("Axe X")
axe.set_ylabel("Axe Y")
axe.set_zlabel("Axe Z")
axe.set_title("Nuage des points")
# Affichage du graphique
plt.show()
```

**Remarques** :

- **fig** : est un objet qui est la figure sur laquelle seront affichées les différents graphiques
- **axe** : est un subplot (sous graphique de la figure
  - **1, 1, 1** : est l'argument qui indique que la figure sera divisée en sous-zone : 1 ligne; 1 colonne et sera en position 1
  - **projection = '3d'** permet la création d'un graphique 3d
    

### Division de la figure en plusieurs **subplot** (2 zones : gauche et droite)

```python
# Création de la figure
fig = plt.figure(figsize = (10,12))
# figure 1, à gauche
axe1 = fig.add_subplot(1, 2, 1, projection = '3d')
# Génération des données aléatoires (20 nombres pour chaque vecteur)
X = np.random.randint(-10, 10, 20)
Y = np.random.randint(-10, 10, 20)
Z = np.random.randint(-10, 10, 20)
# Dessin du nuage des points
axe1.scatter(X, Y, Z, color = "green", marker = "s")
# Titres des axes
axe1.set_xlabel("Axe X")
axe1.set_ylabel("Axe Y")
axe1.set_title("Nuage des points")
# figure 2, à droite (pas de projection 3d)
axe2 = fig.add_subplot(1, 2, 2)
# Génération des données aléatoires (20 nombres pour chaque vecteur)
A = np.linspace(-10, 10, 50)
# Dessin de la courbe représentative de la fonction Tangente Hyperbolique
axe2.plot(A, np.tanh(A), color = 'cyan', linestyle = 'solid')
# Titres des axes
axe2.set_xlabel("Axe X")
axe2.set_title("Fonction Tangente Hyperbolique")
# Ajustement de la mise en page
plt.tight_layout()
# Aspect de l'axe1 : même ratio pour les axes X, Y et Z pour un meilleur rendu
axe1.set_box_aspect([1, 1, 1])
# pour que les deux figures ont une échelle uniforme.
axe2.set_box_aspect(1)
# Affichage du graphique
plt.show()
```

# Surface représentative d'une fonction de deux variables

```python
# les intervalles des abscisses (x) et ordonnées (y)
x = np.linspace(-5, 5, 50)  # Génère 50 valeurs entre -5 et 5 pour l'axe X
y = np.linspace(-5, 5, 50)  # Génère 50 valeurs entre -5 et 5 pour l'axe Y
X, Y = np.meshgrid(x, y)  # Crée une grille 2D de points (X, Y)
# les lignes des X seront constituées du tableau x
# Les colonnes de Y seront constituées du tableau y

# Définition de la fonction à représenter (exemple : f(x, y) = x² + y²)
Z = X**2 + Y**2  # Calcule les valeurs de Z pour chaque point (X, Y)

# Cré ation de la figure et de l'axe 3D
fig = plt.figure(figsize=(10, 10))  # Définit la taille de la figure
axe = fig.add_subplot(1, 1, 1, projection='3d')  # Ajoute un axe 3D à la figure

# Tracé de la surface
axe.plot_surface(X, Y, Z, cmap='viridis')  # Affiche la surface en utilisant un dégradé de couleurs

# Ajout des titres
axe.set_xlabel("Axe X")  # Nom de l'axe X
axe.set_ylabel("Axe Y")  # Nom de l'axe Y
axe.set_zlabel("Axe Z = f(X,Y)")  # Nom de l'axe Z
axe.set_title("Surface de la fonction f(x, y) = x² + y²")  # Titre du graphique

# Affichage du graphique
plt.tight_layout()
plt.show() 
```

# Les diagrammes circulaires (**Camembert**) : 
avec ***plt.pie(valeurs, titre)***, où le premier arguments est une liste des nombres que l'on souhaite visualiser et le second est l'étiquette de chaque nombre (les deux arguments sont donc des listes uni-dimensionnelles de mêmes tailles) 

```python
valeurs = np.random.randint(0, 100, (3))
titre = [" Moins de 18 ans", " 18 - 34 ans", ">= 35 ans"]
plt.pie(valeurs, labels = titre)
plt.show()
```

# Les diagrammes à bar (bâtons) : 
Utilisés pour représenter des variables **catégorielles** (qualitatives)

```python
# Données : catégories et valeurs
categories = ["A", "B", "C", "D"]
valeurs = [10, 15, 7, 12]
# Affichage du diagramme à barres 
plt.bar(categories, valeurs, color='skyblue', edgecolor='black', width=0.3)
# edgecolor est la couleur des contours des bars
# edgewidth est la largeur (épaisseur) des bars
# Ajout des labels et du titre
plt.xlabel("Catégories")
plt.ylabel("Valeurs")
plt.title("Diagramme à Barres")
# Affichage du graphique
plt.show() 
```
